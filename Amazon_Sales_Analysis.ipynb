```python
import pandas as pd
from sqlalchemy import create_engine
df=pd.read_csv('Amazon_Cleaned_Report.csv',low_memory=False)

# clean columns names:
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('-','_')
df.columns = df.columns.str.replace(':','')

# make an engine that connects pthon with mysql
engine = create_engine('mysql+pymysql://root:ShubhData%40161625@localhost:3306/salesdb')
df.to_sql(name='sales_data',con=engine,if_exists='replace',index=False)
```
